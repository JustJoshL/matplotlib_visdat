## 6. Belajar Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')    
plt.title('Contoh Bar Plot')

plt.show()
```

### Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, 
        data,
        color='red',
        alpha=0.75)

plt.grid(linestyle='--', #bisa juga '-', '-.', ':'
         linewidth=4,
         axis='both', #untuk garis" ada di pilihan axis: x, y, both
         alpha=0.25)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')    
plt.title('Contoh Bar Plot')
plt.show()
```

### Horisontal Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
# plt.barh(kategori, data)
plt.barh(kategori[::-1], data[::-1]) #membalik urutan kategori dan jumlah

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')    
plt.title('Contoh Bar Plot')
plt.show()
```

### Stacked Bar Plot

```python
data1= [25, 85, 75, 48, 60]
data2= [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
# plt.bar(kategori, data2, label='Data 2') #kalo tanpa bottom, nanti si bar plotnya bertumpuk, data 2 akan nimpa data 1
plt.bar(kategori, data2, label='Data 2', bottom=data1) #untuk stacked bar plot

plt.legend() #berasosiasi dengan label di atas

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')
plt.show()
```

### Grouped Bar Plot

```python
data1= [25, 85, 75, 48, 60]
data2= [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))  #lokasi labelnya/index kategori
x
```

```python
width = 0.25  #lebar bar, persentase dari jarak antar x

#barplot akan dikategorikan sesuai kategori, dan bersampingan
plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)  #mengganti label x sesuai kategori (ABCDE)

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

## 7. Belajar Pie Plot

Pie Plot Simple

```python
#menggunakan banyak pyplot
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data,               #membuat pie plot, pake fungsi pie.
        labels=kategori,        #opsional, ga harus ada.
        autopct='%1.1f%%', #opsional, menampilkan persentase di tiap bagian pie
        startangle=90)    #memulai pie dari sudut 90 derajat (posisi jam 3)

plt.title('Contoh Pie Plot')
plt.show()
```

Pengaturan Pie Plot Legend dan Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,               
        labels=kategori,   
        colors=warna,     
        autopct='%1.1f%%', 
        startangle=90)   

plt.legend()
plt.title('Contoh Pie Plot')
plt.show()
```

Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = (0., 0., 0.1, 0., 0.08)  #menggeser irisan pertama sejauh 0.1 dari pusat pie
```

```python
plt.pie(data,               
        labels=kategori,   
        colors=warna,     
        explode=explode_var,
        autopct='%1.1f%%', 
        startangle=90)   

plt.legend()
plt.title('Contoh Exploded Pie Plot')
plt.show()
```

## 8. Table Plot/Table Chart

```python
data_uas = [
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['Karti', 71]
]

data_uas
```

OO Style

```python
fig, ax = plt.subplots()
table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(18) #wajib, biar bisa diatur ukuran fontnya
table.scale(0.5, 3)  #mengatur skala ukuran tabel, ukuran kolom, baris

ax.axis(False) #disimpan di dalam axis, dibikin false biar ga ada sumbunya
plt.show()
```

Pyplot style

```python
#berbeda di plottingnya aja
table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4)

ax = plt.gca() #figure dan axis tidak dideklarasikan secara eksplisit, jadinya harus dipanggil gca (get current axis) - mendisable axis
ax.axis(False)

plt.show()
```
