```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Simple line plot

```python
x = np.arange(0.0, 2.0, 0.01) #ada 3 buah parameter: start, stop, step
x #ini inklusif, dan 2.0nya eksklusif
```

```python
s = np.sin(2 * np.pi * x) #trigonometri, nilai radian.
s
```

```python
fig, ax = plt.subplots() #ini yang OO style
ax.plot(x, s)
ax.set(xlabel='nilai x',
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid()

plt.show()
```

```python
#Pyplot style:
plt.plot(x, s)
plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')
plt.grid()
plt.show()
```

### Multiple Subplots

```python
x1 = np.linspace(0.0, 5.0, 100) #100 deret bilangan, dari 0 sampai 5 (inlklusif semuanya)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

#### Dengan OO style

```python
fig, (ax1, ax2) = plt.subplots(2, 1) #2 baris, 1 kolom
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')

plt.show()
```

```python
plt.subplot(2, 1, 1) #2 baris, 1 kolom, plot/access ke-1
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots - Pyplot style yey')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2) #2 baris, 1 kolom, plot/access ke-2
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

## 4. Membuat Histogram

```python
mu, sigma = 100, 15 #mu = mean, sigma = stddev
x = mu + sigma * np.random.randn(10000) #randn = normal distribution 10000 = jumlah data point
x
```

```python
x.shape
```

### Pyplot style

```python
plt.hist(x, bins=50, facecolor='g', alpha=0.50) #bins = jumlah bar/penggabungan data, alpha itu transparansi
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$') #45, 500 adalah posisi koordinat x,y, dan \mu \sigma memunculkan simbolnya
plt.grid()

plt.show()
```

### OO Style

```python
fig, ax = plt.subplots()

ax.hist(x, 
        bins=50, 
        facecolor='r', 
        alpha=0.25)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram - OO Style')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
