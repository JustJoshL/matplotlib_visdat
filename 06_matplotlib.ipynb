## 10. Belajar Box Plot

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show() 
```

```python
plt.boxplot(data, labels=['Data'], showfliers=True, flierprops={'marker':'s', 'markerfacecolor':'g'})
plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4'] 
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plots')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## 11. Belajar Violin Plot

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data, showextrema=True, showmeans=False, showmedians=False, quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data, vert=False)
plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

## 12. Belajar Twin Axes

```python
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp') #exponensial
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('Exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro--', label='log') #logaritma
ax2.set_ylabel('Log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='lower left')

plt.show()
```

```python
x = np.random.randint(16, 40, size=30)
x
```

```python
def C2F(celsius=0):
    return (celsius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect("ylim_changed", konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius (C)')
ax2.set_ylabel('Fahrenheit (F)')

fig.suptitle('Temperatur Udara')
plt.show()
```
